## Fixed precision math helpers

```python
def as_bin(num):
    return format(num, 'b').zfill(8)
```

```python
def fixed_8_5(strBin):
    result = 0
    
    for i in range(0, 5):
        result += int(strBin[7 - i]) * 1/(2**(5-i))
    
    result += int(strBin[1:3], 2)
    
    if strBin[0] == '1':
        result += -4
    
    return result
    
def increment_bin(num):
    num = int(num, 2) + 1
    return format(num, 'b').zfill(8)
    
    
```
